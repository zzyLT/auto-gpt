```python
labels = '专业素养 专业技能 通用能力'.split(' ')
stopwords = '注意 以上 希望 继续 技能关键词'.split(' ')


def is_label(label, text):
                for l in labels:
                    if l != label and l in text:
                        return True
                return False
            
def is_answer(df, ind):
    for stp in stopwords:
        if stp in df[ind]:
            return False
    return True
    
def clean_text(text):
    cleanwords='1、2、3、4、5、6、7、8、9、0、. 、.'.split('、')
    for clw in cleanwords:
        text = text.replace(clw,'')
    return text
for label in labels:
    data[label] = ''

for id, row in data.iterrows():
    df = data['answer'][id].replace('\r','')
    df = df.split('\n')
    print(f'================={id}====================')
    for index, context in enumerate(df):
        count = 0
        for label in labels:
            if label in context:
                count += 1
        if count == 1:
            for label in labels:
                if label in context:
                    text = ''
                    ind = index + 1
                    while ind < len(df) and (not is_label(label, df[ind])):
                        """
                        TODO:
                        清洗回答的内容，去除废话以及重复的回答
                        """
                        if is_answer(df, ind):
                        
                            if ((ind < len(df)-1) and clean_text(df[ind]) == clean_text(df[ind+1])[:len(df[ind])]) or ((ind < len(df)-2) and clean_text(df[ind]) == clean_text(df[ind+2])[:len(df[ind])]):
                                print(f'>>> fait l\'attetion a index {ind}: {df[ind]}')
                                pass
                            elif (ind < len(df)-1) and df[ind][-1] == df[ind+1][0]:
                                text += clean_text(df[ind])[:-1] + clean_text(df[ind+1])
                                print('-- ', df[ind], df[ind+1], 'to--> ', df[ind][:-1] + df[ind+1])
                                ind += 1
                            else:
                                if '：' in df[ind]:
                                    text+=clean_text(df[ind]) if text=='' else f'\n{clean_text(df[ind])}'
                                else:
                                    print('++  ',df[ind-1], '||  ', df[ind])
                                    text+=clean_text(df[ind])
                       
                            
                        ind += 1
                    print(text)
                    data[label][id] = text if data[label][id]=='' else data[label][id] +'\n'+ text
# data
```
