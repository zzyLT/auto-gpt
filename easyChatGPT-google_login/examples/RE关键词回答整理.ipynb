```python
stopwords = '注意 以上 希望 继续 技能关键词 （续） 如下'.split(' ')
p = re.compile(r'[（](.*?)[）]', re.S)


def is_label(label, text, name):
#                 if name in text and '能力' in text and label == '专业技能':
#                     return True
    if len(text.split('：')) > 1:
        for l in labels:
            if l != label and l in text and text.split('：')[1] == '':
                return True
    return False
            
def is_answer(df, ind):
    for stp in stopwords:
        if stp in df[ind]:
            return False
    return True
    
def clean_text(text):
    cleanwords='1、2、3、4、5、6、7、8、9、0、. 、.、-'.split('、')
    if text[-1] == '等':
        text = text[:-1]
    for clw in cleanwords:
        text = text.replace(clw,'')
    return text

def clea_df(data):
    # 整理回答的具体内容 去重及统一格式
    entities = []
    for index, row in data.iterrows():
        for label in labels:
            text = row[label]
            if text != '':
                info = {}
                for t in text.split('\n'):
                    if t[-1:] == '等' or t[-1:] == '。': 
                        t = t[:-1]
                    if t.split('：')[0] in list(info.keys()) and info[t.split('：')[0]]!=None:
                        info[t.split('：')[0]].extend(re.split("(?:[、,，])", t.split('：')[1]) if len(t.split('：')) > 1 else None)
                    else: info[t.split('：')[0]] = re.split("(?:[、,，])", t.split('：')[1]) if len(t.split('：')) > 1 else None
                newT = []
                for key in info:
                    title = key
                    if len(key) > 5 and ((key[-2:] == '能力') or (key[-2:] == '技巧') or (key[-2:] == '技能')):
                        title = key[:-2]
                    entities.append(title)
                    try:
                        entities.extend(info[key])
                        newT.append(f"{title}：{'、'.join(list(set(info[key])))}")
                    except:
                        newT.append(f"{title}")
                data[label][index] = '\n'.join(list(set(newT)))
    entities = pd.DataFrame(list(set(entities)), columns=['entity'])
    return data, entities

def get_data(data):
    
    for label in labels:
        data[label] = ''


    for id, row in data.iterrows():
        df = data['answer'][id].replace('\r','')
        df = data['answer'][id].replace('\r','')
        df = df.split('\n')
        print(df)
        print(f'================={id}====================')
        for index, context in enumerate(df):
            if is_answer(df, index):
                print(f'>> cleaning [[{df[index]}]]')
                count = 0
                for label in labels:
                    if label in context:
                        count += 1
                if count == 1:
                    for label in labels:
                        if label in context and context.split('：')[1] == '':
                            print('is label:', context, len(context.split('：')), context.split('：')[1] == '')
                            text = ''
                            ind = index + 1
                            while ind < len(df) and (not is_label(label, df[ind], row['name'])):
                                """
                                TODO:
                                清洗回答的内容，去除废话以及重复的回答
                                """
                                print(f">> text : {df[ind]}")
                                if is_answer(df, ind) and (not df[ind] in text):

                                    df[ind] = re.sub(p,'',df[ind]).lower()

                                    if ((ind < len(df)-1) and clean_text(df[ind]) == clean_text(df[ind+1])[:len(df[ind])]) or ((ind < len(df)-2) and clean_text(df[ind]) == clean_text(df[ind+2])[:len(df[ind])]):
                                        print(f'>>> fait l\'attetion a index {ind}: {df[ind]}')
                                        pass
                                    elif (ind < len(df)-1) and df[ind][-1] == df[ind+1][0] and len(re.findall('：',df[ind][:-1] + df[ind+1]))<2:
                                        text += clean_text(df[ind])[:-1] + clean_text(df[ind+1])
                                        print('-- ', df[ind], df[ind+1], 'to--> ', df[ind][:-1] + df[ind+1])
                                        ind += 1
                                    else:
                                        if '：' in df[ind]:
                                            text+=clean_text(df[ind]) if text=='' else f'\n{clean_text(df[ind])}'
                                        else:
                                            print('++  ',df[ind-1], '||  ', df[ind])
                                            text+=clean_text(df[ind]) if '：' in df[ind-1] else '\n'+clean_text(df[ind])


                                ind += 1
                            print(text)
                            data[label][id] = text if data[label][id]=='' else data[label][id] +'\n'+ text
        
        
    return clea_df(data)

data, data_entities = get_data(data)
# tmp, tmp_entities = get_data(tmp)
```
