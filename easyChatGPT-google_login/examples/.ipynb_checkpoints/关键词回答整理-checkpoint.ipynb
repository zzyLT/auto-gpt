```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('answer_40posts_05-318continue.csv')
data
```

```python
labels = '专业素养 专业技能 通用能力'.split(' ')
def is_label(label, text):
                for l in labels:
                    if l != label and l in text:
                        return True
                return False
            
for label in labels:
    data[label] = np.nan

for id, row in data.iterrows():
    df = data['answer'][id].replace('\r','')
    df = df.split('\n')
    for index, context in enumerate(df):
        count = 0
        for label in labels:
            if label in context:
                count += 1
        if count == 1:
            for label in labels:
                if label in context:
                    print(label)
                    text = ''
                    ind = index + 1
                    while ind < len(df) and (not is_label(label, df[ind])):
                        """
                        TODO:
                        清洗回答的内容，去除废话以及重复的回答
                        """
                        text+=f'\n{df[ind]}'
                        ind += 1
    #                 print(text)
                    data[label][id] = text[2:]
data
```
